The goal of PCA is to extract information while reducing the number of features
from a dataset by identifying which existing features relate to another. The crux of the algorithm is trying to determine the relationship between existing features, called principal components, and then quantifying how relevant these principal components are. The principal components are used to transform the high dimensional data to a lower dimensional data while preserving as much information. For a principal component to be relevant, it needs to capture information about the features. We can determine the relationships between features using covariance.

```python
#Import necessary package
import numpy as np
```

```python

data = np.array([
    [   1,   2,  -1,   4,  10],
    [   3,  -3,  -3,  12, -15],
    [   2,   1,  -2,   4,   5],
    [   5,   1,  -5,  10,   5],
    [   2,   3,  -3,   5,  12],
    [   4,   0,  -3,  16,   2],
])
```

### Step 1: Standardize the Data along the Features

![image.png](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQLxe5VYCBsaZddkkTZlCY24Yov4JJD4-ArTA&usqp=CAU)




Explain why we need to handle the data on the same scale.

**[TO DO: Insert Answer here]**

```python
standardized_data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
print(standardized_data)
```

![cov matrix.webp](https://dmitry.ai/uploads/default/original/1X/9bd2851674ebb55e404cc3ff5e2ffe65b42ff460.png)

We use the pair - wise covariance of the different features to determine how they relate to each other. With these covariances, our goal is to group / cluster based on similar patterns. Intuitively, we can relate features if they have similar covariances with other features.

### Step 2: Calculate the Covariance Matrix



```python
cov_matrix = np.cov(standardized_data, ddof=0, rowvar=False)
#Print Covariance matrix
print(cov_matrix)
```

### Step 3: Eigendecomposition on the Covariance Matrix


```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

### Step 4: Sort the Principal Components
# np.argsort can only provide lowest to highest; use [::-1] to reverse the list

```python
# np.argsort can only provide lowest to highest; use [::-1] to reverse the list

order_of_importance = np.argsort(eigenvalues)[::-1]
print ( 'The order of importance is :\n {}'.format(order_of_importance))

# utilize the sort order to sort eigenvalues and eigenvectors
sorted_eigenvalues = eigenvalues[order_of_importance]

print('\n\n Sorted eigen values:\n{}'.format(sorted_eigenvalues))
sorted_eigenvectors = eigenvectors[order_of_importance] # sort the columns
print('\n\n The sorted eigen vector matrix is: \n {}'.format(sorted_eigenvectors))
```

Question:

1. Why do we order eigen values and eigen vectors?

- This is because eigenvectors constitute the principal components in a PCA and thus the most important ones will have higher eigenvalues. Thus, ordering them is important so that we tend to proritize the most significant components and their associated eigenvalues as this will make the PCA process much more easier.
2. Is it true we would consider the lowest eigen value compared to the highest? Defend your answer

- No, it is not true. In PCA, we consider the highest eigenvalues because they correspond to the most significant principal components. These principal components capture more variance in the data, making them more informative and essential for understanding the underlying structure of the dataset. The higher eigenvalues are associated with directions in the feature space where the data varies the most, thus containing the most critical information we need to focus on.


You want to see what percentage of information each eigen value holds. You would have print out the percentage of each eigen value using the formula



> (sorted eigen values / sum of all sorted eigen values) * 100



```python
# use sorted_eigenvalues to ensure the explained variances correspond to the eigenvectors

explained_variance = sorted_eigenvalues / np.sum(sorted_eigenvalues) * 100
explained_variance =["{:.2f}%".format(value) for value in explained_variance]
print( explained_variance)
```

# *What are 2 positive effects and 2 negative effects of PCA

Give 2 Benefits and 2 limitations

**BENEFITS**

1) Dimensionality Reduction: PCA can significantly reduce the dimensionality of a dataset by projecting the data onto a lower-dimensional subspace while retaining most of the important information or variance present in the original data. This can lead to faster computation times, improved model performance, and better data visualization.

2) Noise Reduction: PCA can help in separating the signal (important information) from the noise (irrelevant or redundant information) in the data. By retaining only the principal components that capture the most variance, PCA effectively filters out noise and redundant information, leading to a more robust and interpretable representation of the data.

**LIMITATIONS**

1) Loss of Information: While PCA aims to retain most of the important information in the data, some information is inevitably lost during the dimensionality reduction process. The principal components are linear combinations of the original features, and any non-linear relationships or structures in the data may not be captured adequately.

2) Interpretation Challenges: The principal components obtained from PCA are often difficult to interpret, as they are combinations of the original features. Understanding the meaning or significance of each principal component can be challenging, especially in high-dimensional datasets with many features. This can make it difficult to derive insights or communicate the results to stakeholders.

Benefitsof PCA:

Reduces Dimensionality: PCA can reduce the number of features in a dataset while retaining the most important variance, making the dataset smaller and easier to work with. This reduces the complexity of machine learning models, speeds up training, and can prevent the curse of dimensionality, which often causes poor model performance when there are too many features.

Improves Interpretability: By reducing the dataset to a smaller set of uncorrelated principal components, PCA can reveal the underlying structure of the data. This can make it easier to interpret and visualize relationships between data points.

Limitations of PCA:

Loss of Interpretability: While PCA simplifies data by transforming it into a set of principal components, the new components are often difficult to interpret in terms of the original variables. This makes it harder to understand what each principal component represents in real-world terms.

Risk of Information Loss:  PCA removes the least important components (those with lower variance), but these components can sometimes contain useful information. Especially if too many components are discarded, there’s a risk of losing relevant details, which can negatively impact the model's accuracy.
